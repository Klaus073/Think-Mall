```python
import json
json_data = {}

# Initialize a variable to keep track of the last non-NaN category
last_category = None

# Initialize a variable to keep track of the last non-NaN shop name
last_shop_name = None

# Iterate through each row in the DataFrame
for index, row in df_sheet2.iterrows():
    category = row["Category"]
    name = row["Name"]

    # Check if the category is NaN, use the last non-NaN category
    if pd.isna(category) and last_category is not None:
        category = last_category

    # Check if the name is NaN, use the last non-NaN shop name
    if pd.isna(name) and last_shop_name is not None:
        name = last_shop_name

    # Check if the category is not NaN
    if pd.notna(category):
        # Update the last non-NaN category
        last_category = category

        # Check if the category is not in the JSON structure
        if category not in json_data:
            json_data[category] = {}

        # Check if the name is not NaN
        if pd.notna(name):
            # Update the last non-NaN shop name
            last_shop_name = name

            # Create a dictionary for each shop
            shop_data = {
                "Working Hours": row["Working Hours"],
                "Contact Number": row["Contact Number"],
                "Location": row["Location"],
                "Description": row["Description"],
                "Similar Brands": row["Similar Brands"],
                "Other Info": row["Other Info"]
            }

            # Add the shop data to the category's dictionary under the shop name
            json_data[category][name] = shop_data

# Convert the final dictionary to JSON
json_result = json.dumps(json_data, indent=2)

# Display the JSON result
print(json_result)
```

```python
def dataframe_to_json(df):
    json_data = {}
    last_category = None

# Initialize a variable to keep track of the last non-NaN shop name
    last_shop_name = None

    # Iterate through each row in the DataFrame
    for index, row in df.iterrows():
        category = row["Category"]
        name = row["Name"]

        # Check if the category is NaN, use the last non-NaN category
        if pd.isna(category) and last_category is not None:
            category = last_category

        # Check if the name is NaN, use the last non-NaN shop name
        if pd.isna(name) and last_shop_name is not None:
            name = last_shop_name

        # Check if the category is not NaN
        if pd.notna(category):
            # Update the last non-NaN category
            last_category = category

            # Check if the category is not in the JSON structure
            if category not in json_data:
                json_data[category] = {}

            # Check if the name is not NaN
            if pd.notna(name):
                # Update the last non-NaN shop name
                last_shop_name = name

                # Create a dictionary for each shop
                shop_data = {
                    "Working Hours": row["Working Hours"],
                    "Contact Number": row["Contact Number"],
                    "Location": row["Location"],
                    "Description": row["Description"],
                    "Similar Brands": row["Similar Brands"],
                    "Other Info": row["Other Info"]
                }

                # Add the shop data to the category's dictionary under the shop name
                json_data[category][name] = shop_data

    return json_data

# Convert each DataFrame to JSON

json_sheet3 = dataframe_to_json(df_sheet2)


# Combine the JSON structures without a key
combined_json = { **json_sheet3}

# Convert the final dictionary to JSON
json_result = json.dumps(combined_json, indent=2)

# Save the JSON to a file
output_file_path = 'Think Mall/dining.json'
with open(output_file_path, 'w') as json_file:
    json_file.write(json_result)

print(f"Combined JSON structure saved to: {output_file_path}")
```
